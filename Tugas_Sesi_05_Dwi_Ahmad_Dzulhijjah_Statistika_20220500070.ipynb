```python
import matplotlib.pyplot as plt

# Data dari tabel
perusahaan = [
    'Produk Bangunan Hoden',
    'J&R Printing Inc.',
    'Konstruksi Beton',
    'Mancell Heating dan Plumbing',
    'Air Conditioning',
    'Lembaran Logam'
]

penjualan = [1645.2, 4757.0, 8913.0, 627.1, 24612.0, 191.9]

# Membuat grafik batang
plt.figure(figsize=(10, 6))
plt.barh(perusahaan, penjualan, color='skyblue')

# Menambahkan judul dan label sumbu
plt.title('Penjualan Perusahaan Tahun 2019 (dalam ribuan dolar)')
plt.xlabel('Penjualan (ribuan dolar)')
plt.ylabel('Perusahaan')

# Menampilkan plot
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Soal 2

### Kode

```python
import numpy as np
import pandas as pd

# Data produksi dompet
data = [27, 27, 27, 28, 25, 27, 25, 25, 28, 26, 28, 26, 28, 31, 30, 26, 26]

# Menghitung jumlah kelas menggunakan aturan Sturges
n = len(data)
k = int(1 + 3.322 * np.log10(n))

# Menentukan batas bawah dan interval kelas
min_value = min(data)
max_value = max(data)
range_value = max_value - min_value
interval = int(np.ceil(range_value / k))

# Membuat bin atau interval kelas
bins = np.arange(min_value, max_value + interval + 1, interval)

# Membuat tabel distribusi frekuensi
frequencies, bins = np.histogram(data, bins=bins)

# Membuat DataFrame untuk penyajian tabel
df = pd.DataFrame({
    'Kelas Interval': [f'{int(bins[i])} - {int(bins[i+1]-1)}' for i in range(len(bins)-1)],
    'Frekuensi': frequencies
})
```

```python
print(df)
```

---

### **Komentar**

---

> Dari tabel distribusi frekuensi, terlihat bahwa mayoritas produksi dompet terjadi dalam interval 27-28 buah per hari, dengan jumlah yang sedikit pada interval yang lebih tinggi. Hal ini menunjukkan bahwa produksi UMKM tersebut relatif konsisten pada kisaran 27-28 dompet per hari, dengan beberapa hari yang lebih produktif atau kurang produktif. Tabel ini membantu dalam memahami distribusi produksi harian selama periode yang dianalisis.

### Kode Python

```python
import matplotlib.pyplot as plt

# Data dari tabel
nilai_tengah = [150, 210, 270, 330, 390, 450]
frekuensi = [4, 8, 14, 18, 8, 13]

# Membuat grafik polygon
plt.figure(figsize=(10, 6))
plt.plot(nilai_tengah, frekuensi, marker='o', linestyle='-', color='blue')

# Menambahkan judul dan label sumbu
plt.title('Frequency Polygon of Product Prices')
plt.xlabel('Nilai Tengah (S)')
plt.ylabel('Frekuensi')

# Menampilkan plot
plt.grid(True)
plt.show()
```

Jawab :

a. Sebutkan nama bagan tersebut? Bagan yang ditampilkan adalah histogram. Histogram digunakan untuk menyajikan distribusi frekuensi data numerik dalam interval kelas.

b. Berapa jumlah frekuensi dan interval kelasnya?
Jumlah frekuensi adalah total dari semua angka yang ada pada sumbu y (Hari).
Dari gambar:

Interval 5-10: 5 hari
Interval 10-15: 13 hari
Interval 15-20: 28 hari
Interval 20-25: 23 hari
Interval 25-30: 18 hari
Interval 30-35: 10 hari
Interval 35-40: 3 hari

Jumlah frekuensi = 5 + 13 + 28 + 29 + 18 + 10 + 3 = 103 hari

Interval kelasnya adalah jarak antara batas bawah dan batas atas dari setiap kelas, yang dapat dilihat sebagai 5 (contoh: 5-10, 10-15, dan seterusnya).
